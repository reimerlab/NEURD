```python
from datasci_tools import pandas_utils as pu
from pathlib import Path
from mesh_tools import trimesh_utils as tu

def save_synapses_csv_from_segment_id(
    segment_id,
    save_directory = "./",
    filename = None,
    ):
    df = hdju.synapse_df_from_segment_id(segment_id)

    if filename is None:
        filename = f"{segment_id}_synapses.csv"
    filepath = Path(save_directory)  / Path(filename)
    syn_path = pu.df_to_csv(df,f"{str(filepath.absolute())}")
    return syn_path

def save_mesh_from_segment_id(
    segment_id,
    mesh = None,
    suffix = "",
    ):

    if mesh is None:
        mesh = hdju.fetch_segment_id_mesh(segment_id)
    path = tu.write_neuron_off(
        mesh,
        f"{segment_id}{suffix}.off"
    )

    return path

def save_synapses_and_mesh_from_segment_id(
    segment_id,
    verbose = True,
    ):

    syn_path = save_synapses_csv_from_segment_id(
        segment_id = segment_id
    )

    mesh_path = save_mesh_from_segment_id(
        segment_id = segment_id
    )
    
    if verbose:
        print(f"syn_path ={syn_path}")
        print(f"mesh_path ={mesh_path}")
        
    return syn_path,mesh_path

save_synapses_and_mesh_from_segment_id(final_neuron)
```
